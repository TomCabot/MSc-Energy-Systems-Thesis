```python
import pandas as pd
from scipy.stats import pearsonr
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
meaningv1 = pd.read_csv('meaningv1.csv')
survey = pd.read_csv('survey.csv') 
```

```python
#peak = pd.read_csv('pk_watt_byID.csv')
#average = pd.read_csv('avg_watt_byID.csv')
watt_stats = pd.read_csv('watt_stats_byID.csv')
```

```python
# Merge survey and peak.csv's together
survey_watt = pd.merge(watt_stats, survey, on='id')
df1 = survey_watt
```

```python
# Create column 'topPeak' in df1 to show true/false for pk_watt values greater than 1000 
df1['topPeak'] = df1.pk_watt > 1000
```

```python
# removing NaN from 'el_quantile75' column 
df2 = df1[df1['el_quantile75'].notna()]
```

##### USE SAME CODE FOR 'pk_watt' and 'pk/avg watt' WHEN ANALYSING FOR PEARSON COEFFICIENT #######

```python
# Applying Pearson coefficient (p value)
# want p value to be as near to 0 or 1


from scipy.stats import pearsonr

# create empty dataframe
results = pd.DataFrame({
    'key': [], 
    'c': [],
    'p': []
})

for key in df2.keys():              # for each key in all of the df2 keys...1st remove NaN....try 
    df2 = df1[df1[key].notna()]     # removes NaN from df2
    try: 
        c,p = pearsonr(df2['avg_watt'], df2[key])    # SWITCH BETWEEN 'pk_watt' and 'pk/avg watt'
        if p < 0.02:                                # produces the correlation and p value, by comparing 'pk_watt' and all the df2 Keys
            results.loc[len(results.index)]=[key,c,p]
            #print(key,c,p)         # only prints key, c, p if p<0.01
    except:
        pass

results 
    
#results[results.key == 'q9']
```

```python
# Assigning the question meanings to the df. 
# Updated meaning.csv to reflect correct 'keys' to 'meanings' 


df = pd.merge(results, meaningv1, on='key', how='outer')
df = df[['key','question','response','meaning','c','p',]] 
df.dropna(subset=['c','p'],inplace=True)
df
```
